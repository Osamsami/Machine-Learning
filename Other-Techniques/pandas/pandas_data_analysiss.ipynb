##  Step 32: Grouping and Counting with `groupby()` and `size()`

**What is happening here?**
`groupby('city')` splits the DataFrame into groups — one group per unique city. `.size()` counts the number of rows in each group. The result is then sorted in descending order to show which cities hosted the most matches.

**Real-world use:**
`groupby()` is the Pandas equivalent of SQL's `GROUP BY`. It is one of the most powerful tools for aggregating data — counting transactions per store, calculating total sales per product category, or finding average salaries per department. It is used in virtually every data analysis pipeline.

**Purpose in this notebook:**
Reveals that **Mumbai** hosted the most IPL matches (85), followed by Bangalore (66) and Kolkata (61). This reflects historical venue usage patterns in the IPL.

```python
data.groupby('city').size().sort_values(ascending=False)
```

##  Step 33: Grouped Aggregation — Counting Non-Null Values

**What is happening here?**
`data.groupby('city')['winner'].count()` groups by city and then counts only the **non-null** values in the `winner` column for each city. This is slightly different from `.size()`, which counts all rows including nulls. The difference is small here (since `winner` has only 3 null values), but the distinction matters in real data with significant missing values.

**Real-world use:**
Using `.count()` after `groupby()` ensures you're counting actual data, not missing entries. In datasets with many NaN values, `count()` and `size()` can give very different results.

**Purpose in this notebook:**
Confirms match counts per city based on valid (non-null) winner records, showing Mumbai still leads with 85 valid results.

```python
data.groupby('city')['winner'].count().sort_values(ascending=False)
```

## Step 47: Filling Missing Values with `fillna()`

**What is happening here?**
`fillna(0)` replaces **all** `NaN` values in the entire DataFrame with the number `0`. Note that this returns a new DataFrame — the original `data` is unchanged. In the output, `umpire3` now shows `0.0` and missing umpire names show `0` as well.

**Real-world use:**
`fillna()` is a key data imputation technique. Common strategies include filling with `0` (for numeric counts), filling with the mean/median of the column, forward-filling from the previous value (`method='ffill'`), or filling with a specific placeholder string. The right strategy depends on the context.

**Purpose in this notebook:**
Demonstrates the simplest form of missing value imputation — replacing all NaNs with zero. This is appropriate for numeric columns but less appropriate for text columns.

```python
data.fillna(0)
```

Column-specific filling is preferred over whole-DataFrame filling because different columns often need different imputation strategies. For example, numeric columns might be filled with the column mean, while categorical columns might be filled with the mode or a placeholder like `'Unknown'`.

##  Step 50: Filling with a String Value and `inplace=True`

**What is happening here?**
`data['umpire3'].fillna("yes", inplace=True)` fills all NaN values in `umpire3` with the string `"yes"` and **modifies the DataFrame in place** (no new copy is created). A warning is shown by Pandas about incompatible dtypes (float64 vs string), which is expected behavior.

**Real-world use:**
Using `inplace=True` applies the change directly to the existing DataFrame without needing to reassign it. The `inplace` parameter appears throughout Pandas (e.g., `sort_values`, `rename`, `set_index`). In modern best practices, creating a new variable is often preferred over `inplace=True` for clarity and to avoid chained assignment warnings.

**Purpose in this notebook:**
Demonstrates both string filling and the `inplace` modification pattern. After this operation, every row in `umpire3` shows `"yes"` as confirmed in the next cell.

```python
data['umpire3'].fillna("yes",inplace=True)
```
